## understanding the datasets
Here is an explanation of some important columns to the project that will be found in the datasets:Certainly! Here's an explanation of the additional columns you mentioned in the World Happiness Report dataset, including the "sub regions" column:

1. Country name: This column contains the name of each country or region included in the report. It identifies the specific geographical entity for which the happiness scores and other factors are reported.

2. Regional indicator: This column categorizes countries into different regions or geographic areas based on their shared characteristics. It provides an additional level of grouping or classification beyond individual countries. The regional indicator helps in analyzing happiness trends and patterns across different parts of the world.

3. Ladder score: The ladder score column represents the overall happiness score or subjective well-being score of each country. It is a numeric value derived from surveys that ask individuals to rate their life satisfaction or happiness on a scale from 0 to 10. The ladder score provides a standardized measure of happiness, allowing for cross-country comparisons.

4. Logged GDP per capita: This column represents the logarithm of the Gross Domestic Product (GDP) per capita of each country. It measures the economic output per person and indicates the average income or wealth of individuals in that country. Taking the logarithm helps in normalizing the data and addressing the skewed distribution of GDP values.

5. Social support: The social support column quantifies the perceived availability of assistance and support from family, friends, and social networks within a country. It measures the strength of social relationships and the level of social cohesion. Higher values indicate a greater sense of social support and well-being.

6. Healthy life expectancy: This column represents the average life expectancy at birth in each country. It reflects the overall health and well-being of the population and is influenced by factors such as healthcare, sanitation, and lifestyle. Higher values indicate longer life expectancies and better health outcomes.

7. Freedom to make life choices: This column measures the extent to which individuals within a country have the freedom to make life choices. It considers factors such as political and civil liberties, human rights, and individual autonomy. Higher values suggest a greater degree of freedom to choose one's path in life.

8. Generosity: The generosity column quantifies the propensity of individuals within a country to engage in charitable acts and exhibit kindness and benevolence towards others. It measures the willingness to help others and contribute to the well-being of society. Higher values indicate higher levels of generosity.

9. Perceptions of corruption: This column indicates the perceived level of corruption within a country. It measures the extent to which corruption is perceived to exist in public institutions and society as a whole. Higher values suggest a higher perceived level of corruption.

These columns provide insights into various factors that contribute to the overall happiness scores reported in the World Happiness Report. Analyzing these columns helps in understanding the relationship between economic factors, social support, health, freedom, generosity, perceptions of corruption, and happiness levels across different countries and regions.


# Report_2020 data cleaning and preparation


```python
#Adding in the rank column 
Report_2020['Rank'] = Report_2020['Ladder score'].rank(ascending=False).astype(int)
Report_2020.head()
```

```python
#Moving the rank column to the second position column position 
rank_column = Report_2020['Rank']
Report_2020 = Report_2020.drop('Rank', axis=1)
Report_2020.insert(1, 'Rank', rank_column)
Report_2020.head()
```

```python
Report_2020 = Report_2020.rename(columns={'Ladder score' : 'Happiness score'})
```

```python
Report_2020.drop('Regional indicator', axis=1)
```

```python
# Merge the two DataFrames based on the 'Country' column
Report_2020 = pd.merge(Report_2020, region[['name', 'sub-region']], left_on='Country name', right_on='name', how='left', suffixes=('_left', '_right'))

# Drop the redundant 'name' column
Report_2020 = Report_2020.drop('name', axis=1)

# Rename the 'sub-region' column to match the desired column name
Report_2020 = Report_2020.rename(columns={'sub-region_right': 'Subregion'})

# Print the updated DataFrame
Report_2020.columns
```

```python
columns_to_drop = ['Regional indicator']
Report_2020=Report_2020.drop(columns_to_drop, axis=1)
```

```python
#Moving the rank column to the second position column position 
Regional_col = Report_2020['sub-region']
Report_2020 = Report_2020.drop('sub-region', axis=1)
Report_2020.insert(1, 'sub-region',Regional_col)
Report_2020 = Report_2020.rename(columns={'sub-region' : 'Sub-region'})
Report_2020.head()
```

```python
#number of countries grouped per sub region in 2020
Report_2020["Sub-region"].value_counts()
```

```python
#checking for NAN values 
Report_2020.isnull().sum()
```

```python
#dropping NANs values 
Report_2020.dropna(subset=['Sub-region'], inplace=True)
Report_2020.isnull().sum()
```

# # Testing and validation of hypothesis  

```python
# Select the relevant columns for the analysis
X = Report_2020['Logged GDP per capita']  # Independent variable: GDP per capita
y = Report_2020['Happiness score']        # Dependent variable: Happiness score

# Add a constant term to the independent variable (required for the regression model)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Set the level of significance (alpha)
alpha = 0.05

# Print the summary of the regression results
print(model.summary())

# Get the p-value for the coefficient of GDP per capita
p_value = model.pvalues['Logged GDP per capita']

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis. GDP per capita has a significant impact on the happiness score of countries.")
else:
    print("Fail to reject the null hypothesis. GDP per capita has no significant impact on the happiness score of countries.")
```

# Testing and validating hypothesis

```python
# Select the relevant columns for the analysis
X = Report_2021['Logged GDP per capita']  # Independent variable: GDP per capita
y = Report_2021['Happiness score']        # Dependent variable: Happiness score

# Add a constant term to the independent variable (required for the regression model)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Set the level of significance (alpha)
alpha = 0.05

# Print the summary of the regression results
print(model.summary())

# Get the p-value for the coefficient of GDP per capita
p_value = model.pvalues['Logged GDP per capita']

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis. GDP per capita has a significant impact on the happiness score of countries.")
else:
    print("Fail to reject the null hypothesis. GDP per capita has no significant impact on the happiness score of countries.")
```
